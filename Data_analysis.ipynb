# Test the Analysis of variance between AMT_INCOME_TOTAL , AMT_CREDIT and AMT_ANNUITY  

```python
import scipy.stats as stats
stats.f_oneway(dataset["AMT_INCOME_TOTAL"],dataset["AMT_CREDIT"],dataset["AMT_ANNUITY"])
```

# Find the two-way ANOVA

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Two-way ANOVA
ano=ols("AMT_INCOME_TOTAL~TARGET+AMT_CREDIT",data=dataset).fit()
two=sm.stats.anova_lm(ano,type=2)
```
